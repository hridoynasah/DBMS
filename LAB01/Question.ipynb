#### **Question:**

You are tasked with creating a database for a university named `UnivarsityDB`. The database will contain a table called `Students` to store student information. You will perform various operations on this table, including creating the table, inserting data, querying data, modifying the table structure, and performing updates. Write SQL queries to achieve the following tasks.

---

### **Section A: Database and Table Creation**

1. Create a new database named `UnivarsityDB`.

2. Use the `UnivarsityDB` database for all subsequent operations.

3. Create a table named `Students` with the following columns:

   - `StudentID` (INT, Primary Key)
   - `Name` (VARCHAR(50))
   - `Age` (INT, CHECK(Age > 0))
   - `Department` (VARCHAR(30))
   - `CGPA` (DECIMAL(3,2))

4. Display all records from the `Students` table (even if it is empty at this point).

---

### **Section B: Inserting Data**

5. Insert the following student record into the `Students` table:

   - `StudentID`: 1001
   - `Name`: "Ali"

   - `Department`: "CSE"

6. Insert the following multiple student records into the `Students` table:

   - `StudentID`: 1002, `Name`: "Asif", `Age`: 19, `Department`: "Psychology", `CGPA`: 3.80
   - `StudentID`: 1554, `Name`: "Nawshin Zaman", `Age`: 22, `Department`: "CSE", `CGPA`: 3.93
   - `StudentID`: 1003, `Name`: "Babu", `Age`: 24, `Department`: "SWE", `CGPA`: 3.10

7. Insert the following additional student records into the `Students` table:

   - `StudentID`: 1004, `Name`: "ALI KHAN", `Age`: 22, `Department`: "COMPUTER SCIENCE", `CGPA`: 3.80
   - `StudentID`: 1005, `Name`: "SARAH AHMED", `Age`: 21, `Department`: "MATHEMATICS", `CGPA`: 3.65
   - `StudentID`: 1006, `Name`: "JOHN DOE", `Age`: 23, `Department`: "PHYSICS", `CGPA`: 3.45

8. Display all records from the `Students` table after inserting the above data.

### **Section C: Filtering Data**

10. Write an SQL query to display all students who belong to either the "CSE" department or the "Computer Science" department.

11. Write an SQL query to display all students whose CGPA is between 3.5 and 4.0 (inclusive), belong to the "CSE" department, and are aged 25 or younger.
